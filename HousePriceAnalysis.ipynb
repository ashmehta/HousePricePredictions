```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/Users/akshitmehta/Desktop/HousePricePrediction/American_Housing_Data_20231209.csv')
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)  
```

```python
data.isna().sum()  
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=data, x= 'Living Space', y='Price', alpha = 0.5)
plt.title('Price Vs Living Space')
plt.xlabel('Living Space')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=data, x= 'Beds', y='Price', alpha = 0.5)
plt.title('Price Vs Number of Beds')
plt.xlabel('Beds')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=data, x= 'Baths', y='Price', alpha = 0.5)
plt.title('Price Vs Number of Bathrooms')
plt.xlabel('Baths')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(40,16))
sns.barplot(data=data, x= 'City', y='Price')
plt.title('Price Vs City')
plt.xlabel('City')
plt.ylabel('Price')
plt.xticks(rotation=90, fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.barplot(data=data, x= 'State', y='Price')
plt.title('Price Vs State')
plt.xlabel('State')
plt.ylabel('Price')
plt.xticks(rotation=45, fontsize=8)
plt.tight_layout()
plt.show()
```
